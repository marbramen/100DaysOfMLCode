### Linear Regression

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0,9)
y = np.array([19,20,20.5,21.5,22,23,23,25.5,24]).T

plt.plot(x,y,"bo")
plt.grid()
plt.show()
```

```python
A = np.array([np.ones(9), x]).T
w0, w1 = np.linalg.lstsq(A,y,rcond=None)[0]

plt.plot(x, w0 + w1*x, 'r')

for i in range(len(x)):
    plt.plot([x[i], x[i]], [y[i], w1*x[i] + w0], 'g')
plt.show()
```
