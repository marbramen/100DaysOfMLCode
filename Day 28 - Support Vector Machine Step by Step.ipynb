```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

It's important to add a one identical dimension for each x value

```python
# Input data
x = np.array([
    [-2,4,-1],
    [4,1,-1],
    [1, 6, -1],
    [2, 4, -1],
    [6, 2, -1],
])

# Output label
y = np.array([-1,-1,1,1,1])
```

```python
for val, inp in enumerate(x):
    if y[val] == -1:
        plt.scatter(inp[0], inp[1], s=100, marker='*', color='blue',linewidths=1)
    else:
        plt.scatter(inp[0], inp[1], s=100, marker='o', color='red', linewidths=1)
plt.plot()
```

A discriminative classifier can be created using support vector machines. Decision boundary is drawn by maximizing the margin(space) between the line(hyperplane) and classes. Points that are closest to the decision boundary are called support. They are called so because they support the creation of the hyperplane

```python
def svm_function(x,y):
    # initialize weight
    w = np.zeros(len(x[0]))
    l_rate =1 
    out = []
    epoch = 10000
    for e in range(epoch):
        for i, _ in enumerate(x):
            val1= np.dot(x[i], w)
            if (y[i]*val1 < 1):
                w += l_rate*(y[i]*x[i] - 2*w*(1/epoch))
            else:
                w += l_rate*(-2*(1/epoch)*w)
            
        print("Epoch(" + str(e) +"): " + str(np.dot(x,w)))
    for i,val in enumerate(x):
        out.append(np.dot(x[i],w))
    return w,out
```

```python
w, out = svm_function(x,y)
```

```python
print('Calculated weights')
print(w)
```

```python
print('predicte outputs')
print(out)
```
